# Officially running the scrape

```python
import pandas as pd

df = pd.DataFrame(columns=['name', 'description', 'email', 'website'])

df
```

```python
matrix = []

for item in soup.find_all('p')[1:]:
    
    new_entry = []
    
    try:
        new_entry.append(item.find('strong').get_text())
    except:
        try:
            new_entry.append(item.find('b').get_text())
        except:
            new_entry.append(' ')
    
    try:
        new_entry.append(item.find('span').get_text())
    except:
        new_entry.append(' ')
    
    try:
        new_entry.append(item.find('a').get("href")[7:])
    except:
        new_entry.append(' ')
    
    
    try:
        new_entry.append(item.find_all('a')[1].get("href"))
    except:
        new_entry.append("csulb.edu")
                
        
    matrix.append(new_entry)
```

```python
matrix
```

```python
df = pd.DataFrame(matrix, columns=['name', 'description', 'email', 'website'])
df
```
