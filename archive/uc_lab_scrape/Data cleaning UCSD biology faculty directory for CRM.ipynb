```python
import pandas as pd

df = pd.read_csv("ucsd_biology_faculty_directory.txt", sep='\t', lineterminator='\r')

df.head()
```

# Remove O from Office numbers 

```python
df.dtypes
```

```python
for index, row in df.iterrows():
    df.at[index, "Office Number"] = str(df.at[index, "Office Number"])[3:]


df.head()
```

# Remove L from Lab numbers

```python
for index, row in df.iterrows():
    df.at[index, "Lab Number"] = str(df.at[index, "Lab Number"])[3:]


df.head()
```

# Turn lab assistants into their own contact

```python
new_entries = []

for index, row in df.iterrows():

    assistant_value = df.at[index, "Assistant(delete)"]
    if not pd.isnull(assistant_value):
        chopped = [x.strip() for x in str(assistant_value).split(',')]
        
        lab_name = str(df.at[index, "Lab"])
        new_entry = [None, 
                     lab_name, 
                     chopped[0], #assistant's name set under "professor's full name" column
                     "Lab assistant", #assistant's title
                     None, #skip the assistant(delete) column
                     chopped[1], #assistant email
                     None, #skip address
                     chopped[2], #office number
                     None #skip lab number
                    ]
        new_entries.append(new_entry)
        
        
for row in new_entries:
    df.loc[len(df)] = row
```
